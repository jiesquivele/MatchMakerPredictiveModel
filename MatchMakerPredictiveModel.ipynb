```python
X = df[["edhomrec","nachom","grocuhom","iuhom","escivhom"]]
```

```python
y = df[["edmujrec","nacmuj","grocumuj","iumuj","escivmuj"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn import preprocessing 
```

```python
X_train["nachom"][X_train["nachom"]== 531]="999 - Ignorado"
```

```python
X_train=X_train.apply(preprocessing.LabelEncoder().fit_transform)
y_train=y_train.apply(preprocessing.LabelEncoder().fit_transform)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
X_test["nachom"][X_test["nachom"]== 531]="999 - Ignorado"
```

```python
X_test=X_test.apply(preprocessing.LabelEncoder().fit_transform)
```

```python
y_test=y_test.apply(preprocessing.LabelEncoder().fit_transform)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_test
```

```python
y_pred
```

```python
y_pred=pd.DataFrame(y_pred)
y_pred
```

```python
y_pred[0]
```

```python
# Classification results on test set
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test["edmujrec"], y_pred[0])
print("Accuracy: {:.2f}".format(accuracy))

from sklearn.metrics import confusion_matrix, classification_report
cm=confusion_matrix(y_test["edmujrec"],y_pred[0])
print('Confusion Matrix: \n', cm)
print(classification_report(y_test["edmujrec"], y_pred[0]))
```

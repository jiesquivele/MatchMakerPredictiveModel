```python
INTRODUCTION

Context:
    - What it will work for.
Question/Objective:
    - 
Justification:
    - Importance of the tool.

    DATASET DESCRIPTION

    
Description: (***MENOS TABULADO)
    This dataset, by the name "Costa Rica: Total Matrimonios 2017 - 2018", describes Catholic 
    marriages and legal unions among Costa Ricans in both urban and rural areas of the country.
    The main source of the data is from "Registro Civil de Costa Rica"
    
    
    "1. Recepción de material: Funcionarios de la UED se movilizan a las oficinas centrales del Registro Civil, una vez a 
    "la semana, a retirar todos aquellos certificados que han sido enviados por los sacerdotes y notarios públicos."
    
    It was created by Adriana Moya Leiva from the "Instituto Nacional de Estadisticas y Censos" 
    and its "Unidad de Estadistica y Censos". The creation process was as follows: 1) Weekly 
    reception of material from the "Registro Nacional", 2) Organization of material by month,
    3) Coding, 4) Data typing, 5) Control tabulations and 6) Tabulations to be used by the 
    "Centro de informacion". The data was accessed directly from their website at the url:
    "http://sistemas.inec.cr/pad5/index.php/catalog/253#metadata-data_collection". It doesn't
    have a particular license; however, its used is defined by the Law No. 7839 of date 

Name:
    Costa Rica: Total Matrimonios 2017 - 2018
        
Source/Creator
    Instituto Nacional de Estadistica y Censos
    Area de Censos y Encuestas
    Unidad de Estadisticas Demograficas
    Autora: Adriana Moya Leiva
        
Creation Process:
    1) Recepcion semanal de material a partir de los certificados procedentes del Registro Nacional
    2) Ordenamiento mensual del material
    3) Codificacion y critica
    4) Digitacion
    5) Tabulaciones de control
    6) Tabulados para uso del Centro de Informacion
Access Method:
    Descarga directa en la pagina web: http://sistemas.inec.cr/pad5/index.php/catalog/253#metadata-data_collection
    
License:
    Sin licencia???
    
Publication Date/Recovery Date:
    Pulication: June 2, 2020
    Recovery: August 5, 2020
    
File Format: 
    ".sav"
    
Data Dictionary:
anotrab - Año de trabajo
mestrab - Mes de trabajo
clasemat - Clase de matrimonio
provocu - Provincia de ocurrencia
pcocu -Cantón de ocurrencia
pcdocu - Distrito del ocurrencia
mesmat - Mes que se realiza el matrimonio
anomat - Año que se realiza el matrimonio
edadhom - Edad del hombre
edhomrec - Edad del hombre en grupos
paishom - País de origen del hombre
nachom - Nacionalidad del hombre
grocuhom - Ocupación del hombre en grupos
provhom - Provincia de residencia del hombre
pchom - Cantón de residencia del hombre
pcdhom - Distrito de residencia del hombre
iuhom - Índice de urbanidad del hombre
escivhom - Estado conyugal del hombre
numathom - Número de matrimonios del hombre
edadmuj - Edad de la mujer
edmujrec - Edad de la mujer en grupos
paismuj - País de origen de la mujer
nacmuj - Nacionalidad de la mujer
grocumuj - Ocupación de la mujer en grupos
provmuj - Provincia de residencia de la mujer
pcmuj - Cantón de residencia de la mujer
pcdmuj - Distrito de residencia de la mujer
iumuj - Índice de urbanidad de la mujer
escivmuj - Estado conyugal de la mujer
numatmuj - Número de matrimonios de la mujer
diadeclara - Día que se realiza la declaración
mesdeclara - Mes que se realiza la declaración
anodeclara - Año que se realiza la declaración

Source: INEC "http://sistemas.inec.cr/pad5/index.php/catalog/253#metadata-data_collection".
    
```

```python
X = df[["edhomrec","nachom","grocuhom","iuhom","escivhom"]]
```

```python
y = df[["edmujrec","nacmuj","grocumuj","iumuj","escivmuj"]]
```

```python
X["nachom"][X["nachom"]== 531]="999 - Ignorado"
```

```python
from sklearn import preprocessing 
```

```python
X=X.apply(preprocessing.LabelEncoder().fit_transform)
y=y.apply(preprocessing.LabelEncoder().fit_transform)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

# RandomizedSearchCV

```python
# Import necessary modules
from scipy.stats import randint
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

# Setup the parameters and distributions to sample from: param_dist
param_dist = {"max_depth": [3, None],
              "max_features": randint(1, 9),
              "min_samples_leaf": randint(1, 9),
              "criterion": ["gini", "entropy"]}

# Instantiate a Decision Tree classifier: tree
tree = DecisionTreeClassifier()

# Instantiate the RandomizedSearchCV object: tree_cv
tree_cv = RandomizedSearchCV(tree, param_dist, cv=5, scoring="accuracy")

# Fit it to the data
tree_cv.fit(X_train, y_train)

# Print the tuned parameters and score
print("Tuned Decision Tree Parameters: {}".format(tree_cv.best_params_))
print("Best score is {}".format(tree_cv.best_score_))
```
